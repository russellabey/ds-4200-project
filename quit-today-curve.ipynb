Quit-today chart curve from `lung_cancer.csv`


```python
import pandas as pd

df = pd.read_csv("lung_cancer.csv")
smokers = df.loc[df["smoker"] == 1].copy()

n_bins = 15
smokers["py_bin"] = pd.qcut(smokers["pack_years"], q=n_bins, duplicates="drop")
grp = smokers.groupby("py_bin", observed=True)

curve = (
    grp.agg(
        pack_years=("pack_years", "mean"),
        chance_cancer_pct=("lung_cancer_risk", lambda s: 100 * s.mean()),
        n=("pack_years", "size"),
    )
    .reset_index(drop=True)
    .sort_values("pack_years")
    .reset_index(drop=True)
)

curve
```
